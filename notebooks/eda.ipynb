### Important Information for Cleaning the Data: What data do we need and how should we present it?

#### First which headers are essential/important?

When training a model for tasks such as POS tagging, dependency parsing, or other NLP tasks using an annotated corpus like the GUM corpus, several pieces of information are crucial. Here are the key components we could consider:

1. **Tokens**
    - **Description:** The individual words or tokens in the text.
    - **Importance:** Tokens are the basic units of analysis for most NLP tasks. They are the input to your model.

2. **POS Tags (UPOS and XPOS)**
    - **Description:** Part-of-speech tags that provide syntactic information about each token.
    - **Importance:** POS tags are often used as features in various NLP tasks, including dependency parsing and named entity recognition.

3. **Lemmas**
    - **Description:** The base or dictionary form of each token.
    - **Importance:** Lemmas help in reducing the dimensionality of the feature space by grouping different forms of a word together.

4. **Morphological Features (FEATS)**
    - **Description:** Additional grammatical features of tokens, such as tense, number, gender, etc.
    - **Importance:** These features provide more detailed syntactic and semantic information, which can improve model performance.

5. **Dependency Relations (HEAD and DEPREL)**
    - **Description:** Information about the syntactic structure of the sentence, including the head of each token and the type of dependency relation.
    - **Importance:** Dependency relations are crucial for tasks like dependency parsing and can also be useful for other syntactic and semantic tasks.

6. **Named Entities (Entity)**
    - **Description:** Annotations that identify named entities in the text.
    - **Importance:** Named entity recognition (NER) is a common NLP task, and these annotations can be used to train and evaluate NER models.

7. **Coreference Information**
    - **Description:** Annotations that link different mentions of the same entity within the text.
    - **Importance:** Coreference resolution is important for understanding the context and relationships between entities in a text.

8. **Discourse Annotations**
    - **Description:** Information about the discourse structure of the text, such as discourse relations and organization.
    - **Importance:** Discourse annotations can be useful for tasks like text summarization and understanding the overall structure of the text.

9. **Miscellaneous Annotations (MISC)**
    - **Description:** Additional annotations that may include morphological segmentation, construction grammar, information status, Wikification, etc.
    - **Importance:** These annotations provide extra layers of information that can be leveraged for more advanced NLP tasks.

## Baseline/Model Testing Requirements

When performing baseline/model testing for POS tagging, there are several key components and steps we need to consider. These include preparing the data, selecting a baseline model, evaluating the model, and iterating on improvements. Here’s a detailed breakdown:

### 1. **Data Preparation**
- **Tokenization**: Ensure the text is properly tokenized into individual words or tokens.
- **POS Tags**: Extract the POS tags (both UPOS and XPOS) for each token.
- **Train-Test Split**: Split your dataset into training and testing sets to evaluate model performance.

### 2. **Baseline Model Selection**
- **Simple Models**: Start with simple models such as:
  - **Most Frequent Tag**: Assign the most frequent POS tag from the training set to each token in the test set.
  - **Unigram Tagger**: Tag each word based on the most frequent tag observed in the training set.
  - **Bigram/Trigram Tagger**: Use the context of the previous one or two words to predict the tag.

### 3. **Feature Engineering**
- **Basic Features**: Use basic features such as the word itself, prefixes, suffixes, capitalization, etc.
- **Contextual Features**: Include features from surrounding words (e.g., previous and next words).

### 4. **Model Training**
- **Train the Model**: Use the training set to train the baseline model.
- **Hyperparameters**: Set and tune hyperparameters if applicable.

### 5. **Model Evaluation**
- **Accuracy**: Measure the accuracy of the model on the test set.
- **Confusion Matrix**: Analyze the confusion matrix to understand which tags are often confused.
- **Precision, Recall, F1-Score**: Calculate these metrics for a more detailed evaluation.

### 6. **Iteration and Improvement**
- **Error Analysis**: Analyze errors to identify areas for improvement.
- **Feature Addition**: Add more features or use more complex models (e.g., CRFs, HMMs, neural networks).
- **Cross-Validation**: Use cross-validation to ensure the model generalizes well.

```python
####This checks if the tokens and xpos_tags are consistent
from sklearn.model_selection import train_test_split
from collections import Counter
import numpy as np

# Check for mismatches and filter out inconsistent entries
consistent_tokens = []
consistent_xpos_tags = []
consistent_feats = []

for i in range(len(tokens)):
    if len(tokens[i]) == len(xpos_tags[i]):
        consistent_tokens.extend(tokens[i])
        consistent_xpos_tags.extend(xpos_tags[i])
        consistent_feats.extend(feats[i])
    else:
        print(f"Mismatch at index {i}: {len(tokens[i])} tokens, {len(xpos_tags[i])} tags")

print(len(tokens))
# Verify the lengths after filtering
print(f"Number of tokens: {len(consistent_tokens)}")
print(f"Number of XPOS tags: {len(consistent_xpos_tags)}")
print(f"Number of FEATS: {len(consistent_feats)}")

```

```python
# Install the sklearn-crfsuite package
%pip install --upgrade pip
%pip install sklearn-crfsuite

import sklearn_crfsuite
from sklearn_crfsuite import metrics
from sklearn.model_selection import train_test_split
```

```python

# Feature extraction function
def extract_features(tokens, i):
    token = tokens[i]
    features = {
        'bias': 1.0,
        'token.lower()': token.lower(),
        'token.isupper()': token.isupper(),
        'token.istitle()': token.istitle(),
        'token.isdigit()': token.isdigit(),
    }
    if i > 0:
        token1 = tokens[i-1]
        features.update({
            '-1:token.lower()': token1.lower(),
            '-1:token.istitle()': token1.istitle(),
            '-1:token.isupper()': token1.isupper(),
        })
    else:
        features['BOS'] = True

    if i < len(tokens)-1:
        token1 = tokens[i+1]
        features.update({
            '+1:token.lower()': token1.lower(),
            '+1:token.istitle()': token1.istitle(),
            '+1:token.isupper()': token1.isupper(),
        })
    else:
        features['EOS'] = True

    return [features]

def extract_features_for_sentence(tokens):
    return [extract_features(tokens, i) for i in range(len(tokens))]

def get_labels(tags):
    return [tags]
```

BOS and EOS are special features that stand for "Beginning of Sequence" and "End of Sequence," respectively.

```python

# Prepare the data for CRF
x = extract_features_for_sentence(filtered_tokens)
y = [get_labels(tags) for tags in filtered_xpos_tags]

```

```python

for i in range(100):
    if len(x[i]) != len(y[i]):
        print(f"Mismatch at index {i}: {(x[i])} = {len(x[i])}, {(y[i])} = {len(y[i])}")

print(len(x), len(y))
```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train the CRF model
crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,
    c2=0.1,
    max_iterations=100,
    all_possible_transitions=True
)
crf.fit(X_train, y_train)

# Predict on the test set
y_pred = crf.predict(X_test)

# Evaluate the model
labels = list(crf.classes_)
metrics.flat_f1_score(y_test, y_pred, average='weighted', labels=labels)

# Print classification report
print(metrics.flat_classification_report(y_test, y_pred, labels=labels, digits=3))
```

The output is a classification report which evaluates the performance of a Conditional Random Fields (CRF) model on a POS tagging task. The report includes several metrics for each POS tag, as well as overall metrics for the entire dataset.

### Key Metrics in the Classification Report:

1. **Precision**: The ratio of correctly predicted positive observations to the total predicted positives. It is calculated as:
   $$
   \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}
   $$

2. **Recall**: The ratio of correctly predicted positive observations to the all observations in the actual class. It is calculated as:
   $$
   \text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}
   $$

3. **F1-Score**: The weighted average of Precision and Recall. It is calculated as:
   $$
   \text{F1-Score} = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
   $$

4. **Support**: The number of actual occurrences of the class in the dataset.

### Explanation of the Output:

#### Per-Class Metrics:
For each POS tag (e.g., `JJ`, `VBP`, `NN`, etc.), the report provides:
- **Precision**: The precision for the specific POS tag.
- **Recall**: The recall for the specific POS tag.
- **F1-Score**: The F1-score for the specific POS tag.
- **Support**: The number of true instances for the specific POS tag in the test set.

#### Overall Metrics:
- **Micro Average**: Averages the total true positives, false negatives, and false positives. It is useful when you want to know the overall performance of the model.
- **Macro Average**: Averages the metrics for each class without considering the class imbalance. It treats all classes equally.
- **Weighted Average**: Averages the metrics for each class while considering the class imbalance. It gives more weight to classes with more instances.

### Warnings:
The warnings indicate that some labels have no true or predicted samples, which makes precision, recall, and F1-score ill-defined for those labels. These warnings are common in imbalanced datasets where some classes have very few or no instances.




### Key Points:
- **High Precision, Recall, and F1-Score**: For many common POS tags (e.g., `JJ`, `NN`, `VB`), the model performs well with high precision, recall, and F1-score.
- **Low Scores for Rare Tags**: For rare tags (e.g., `AFX`, `ADD`, `NNPS`), the model has lower precision, recall, and F1-score, indicating difficulty in predicting these tags accurately.
- **Warnings**: The warnings indicate that some tags have no true or predicted samples, leading to undefined metrics for those tags.

### How CRF Uses Morphological Features:
The CRF model uses the morphological features provided in the `feats` column as part of the feature set for each token. These features help the model understand the grammatical properties of each token, such as number, degree, tense, etc. By including these features, the model can make more informed predictions based on the additional linguistic information.

By incorporating the `feats` column into the feature extraction process, you can provide the CRF model with richer information about each token, potentially improving the model's performance in sequence labeling tasks like POS tagging.

### Explanation of How CRF Uses Morphological Features:

1. **Feature Extraction**:
    - The `extract_features` function creates a dictionary of features for each token.
    - This dictionary includes basic token features (like lowercase form, whether the token is uppercase, etc.) and morphological features from the `feats` column.
    - The morphological features are added to the `features` dictionary using `features.update(feature_dict)`.

2. **Training the CRF Model**:
    - The CRF model is trained using the feature dictionaries for each token in the training data.
    - **The model learns patterns based on the features provided, including the morphological features**.

3. **Making Predictions**:
    - When making predictions, the CRF model uses the same feature extraction process to create feature dictionaries for each token in the test data.
    - The model then uses these features to predict the POS tags for each token.


To incorporate the `feats` column into the CRF model, you need to ensure that the morphological features from the `feats` column are included in the feature extraction process. The `feats` column contains dictionaries with morphological features for each token, such as `{'Number': 'Sing'}` or `{'Degree': 'Pos'}`.

### Steps to Incorporate the `feats` Column:

1. **Update the Feature Extraction Function**: Modify the feature extraction function to include the morphological features from the `feats` column.
2. **Handle `None` Values**: Ensure that `None` values in the `feats` column are handled appropriately.

### Updated Feature Extraction Function:

Here is the updated feature extraction function that includes the morphological features from the `feats` column:



```python
# Feature extraction function
def extract_features(tokens, feats, i):
    token = tokens[i]
    feature_dict = feats[i] if feats[i] is not None or "" else {}
    features = {
        'bias': 1.0,
        'token.lower()': token.lower(),
        'token.isupper()': token.isupper(),
        'token.istitle()': token.istitle(),
        'token.isdigit()': token.isdigit(),
        'prefix-1': token[:1],
        'prefix-2': token[:2],
        'prefix-3': token[:3],
        'suffix-1': token[-1:],
        'suffix-2': token[-2:],
        'suffix-3': token[-3:],
        'token.length': len(token),
    }
    if i > 0:
        token1 = tokens[i-1]
        features.update({
            '-1:token.lower()': token1.lower(),
            '-1:token.istitle()': token1.istitle(),
            '-1:token.isupper()': token1.isupper(),
        })
    else:
        features['BOS'] = True

    if i < len(tokens)-1:
        token1 = tokens[i+1]
        features.update({
            '+1:token.lower()': token1.lower(),
            '+1:token.istitle()': token1.istitle(),
            '+1:token.isupper()': token1.isupper(),
        })
    else:
        features['EOS'] = True

    features.update(feature_dict)  # Include morphological features

    return [features]


def extract_features_for_sentence(tokens, feats):
    return [extract_features(tokens, feats, i) for i in range(len(tokens))]

def get_labels(tags):
    return [tags]
# Prepare the data for CRF
x = extract_features_for_sentence(filtered_tokens, filtered_feats)
y = [get_labels(tags) for tags in filtered_xpos_tags]

for i in range(10):
    print(x[i])
    print(y[i])
```
